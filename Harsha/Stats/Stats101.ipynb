```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
video6=pd.read_csv('Videos6.csv')
```

```python
video6.head(10)
```

```python
video6.info()
```

```python
from datetime import datetime
```

```python
video6['trending_date']
```

```python
video6['trending_date']=video6['trending_date'].map(lambda x:datetime.strptime(x, '%y.%d.%m'))
#Converting Trending Date to required format
```

```python
video6['publish_time']=video6['publish_time'].map(lambda x: x.split('T')[0]).map(lambda x:datetime.strptime(x, '%Y-%m-%d'))
#Converting Publish Date to required format
```

```python
video6['publish_time']=pd.to_datetime(video6['publish_time'])
```

```python
trending_duration=(video6['trending_date']-video6['publish_time']).dt.days
# Getting No of days a video took to get trended
```

```python
# Calculating the mean and median of No of Days to get trended
print('The average no. of days for a video get trended is {}'.format(np.ceil(trending_duration.mean())))
print('The median of trending days is',(trending_duration).median())
```
