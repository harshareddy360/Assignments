```python
import yfinance as yf
import warnings
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
stocks_names=['BHARTIARTL.NS',"SBIN.NS","DRREDDY.NS","GSPL.NS","JSWSTEEL.NS","JSWENERGY.BO","RELIANCE.NS","YESBANK.NS","TATAMOTORS.NS",
"LAURUSLABS.NS",
"NTPC.NS",
"SAIL.BO"]# List of all the stokes names 
```

```python
stocks=[]
for i in stocks_names:
    data=yf.download(i, start="2015-01-01", end="2022-12-31")
    data=data.reset_index()
    data['stocks']=i
    stocks.append(data)
merged_df = pd.concat(stocks, ignore_index=True) 
# Importing all the data from yfinance and merging all the data in one data set 
```

```python
merged_df=merged_df.drop(columns=['High','Low','Adj Close','Volume']) # Droping all the unwanted columns in the data set 
```

```python
merged_df['percentage']=(merged_df['Close']-merged_df['Open'])/merged_df['Open'] 
# finding the percentage of the profit 
```

```python
final_df=merged_df[merged_df['percentage']>0.053]
# filtering the stocks which has more of profit persentage.
```

```python
final_df=final_df.sort_values('percentage',ascending=False)
```

```python
final_df['Date']=final_df['Date'].drop_duplicates()
final_df.dropna(inplace=True)
#Droping all the duplicates from the dataset
```

```python
final_df=final_df.sort_values('Date')
#sorting the dataset in the decending order
```
